# Regressão e Correlação Linear múltipla

Dada uma amostra $(Y_i, X_{i1}, \dots, X_{ip})\;_{i \; \in \; \{1, \dots, n\}}$ tentamos explicar, com a maior precisão possível, os valores tomados por $Y_i$, denominada variável endógena, de uma série de variáveis explicativas $X_{i1},\dots,X_{ip}$ . O modelo teórico, formulado em termos de variáveis ​​aleatórias, assume a forma:

$$Y_i = a_0 + a_1 X_{i1} + a_2 X_{i2} + \dots + a_p X_{ip} + \epsilon_{i,} \quad \quad i = 1,\dots,n$$

onde $\epsilon_i$ é o erro do modelo que expressa, ou resume, a informação que falta na explicação linear dos valores de $Y_i$ de $X_i,\dots,X_{ip}$ (problema de especificação, variáveis não levadas em consideração, etc.). Os coeficientes $a_0, a_1,\dots,a_p$ são os parâmetros a serem estimados.

## Validando o Modelo

Precisamos ter em mente as premissas/suposições da regressão linear, caso contrário, o modelo falhará em fornecer bons resultados com os dados. Portanto, para um modelo bem-sucedido, é muito crucial validar essas suposições no conjunto de dados fornecido. Vamos tratar das seguintes premissas: 

**1.** O relacionamento entre a variável dependente e independente devem ser linear;
 
 
**2.** Não deve haver correlação entre as variáveis independentes;


**3.** Os resíduos devem ter uma distribuição normal;


**4.** Os resíduos devem ter variância constante (homocedasticidade);


**5.** Os resíduos não devem apresentar auto-correlação;

### 1. Linearidade entre variáveis independentes e dependentes

Podemos analisar essa premissa de duas maneiras: Da forma gráfica e da forma númerica por meio de testes de correlações.

```python
ax = sns.pairplot(colegiox, y_vars='NOTA_REDACAO', x_vars=['COMP1', 'COMP2', 'COMP3', 'COMP4', 'COMP5'])
```

```python
correlacoes = colegiox.corr(method='pearson')
plt.figure(figsize=(14, 11))
sns.heatmap(correlacoes, annot=True)
```

É possível visualizar, pelo heatmap, que existe boas correlações lineares entre a nossa variável dependente e as variáveis independentes. Se utilizarmos a seguinte interpretação do coeficiente: 
<center>$\rho = 1 : \text{Perfeita}$</center>
<center>$0.95 \leq \rho < 1 : \text{Muito Forte}$</center>
<center>$0.8 \leq \rho < 0.95 : \text{Forte}$</center>
<center>$0.5 \leq \rho < 0.8 : \text{Moderada}$</center>
<center>$0 \leq \rho < 0.5 : \text{Fraca}$</center>

Temos: 
<center>$\text{COMP1: Moderada correlação linear}$</center>
<center>$\text{COMP2: Forte correlação linear}$</center>
<center>$\text{COMP3: Forte correlação linear}$</center>
<center>$\text{COMP4: Forte correlação linear}$</center>
<center>$\text{COMP5: Forte correlação linear}$</center>

Apesar da análise gráfica já ter oferecido uma boa resposta quanto a relação linear, o teste numérico é sempre importante, uma vez que uma análise gráfica pode conter conteúdo subjetivo e a numérica não.

### 2. Ausência de Multicolinearidade
Quando as variáveis **independentes** são correlacionadas entre si, temos um problema chamado multicolinearidade.

Isso leva ao desenvolvimento de um modelo com coeficientes que possuem valores que dependem da presença de outras variáveis. Em outras palavras, teremos um modelo que mudará drasticamente se uma variável independente for removida, portanto um modelo como esse será impreciso.

Consideramos multicolinearidade quando:

<center>$0.95 \leq \rho < 1 : \text{Multicolinearidade considerada}$</center>

```python
colegiox[['COMP1','COMP2','COMP3', 'COMP4','COMP5']].corr() >= 0.95
```

Nenhuma variável independente apresenta multicolinearidade.

### 3. Normalidade dos resíduos

Para que a nossa regressão linear funcione de acordo, devemos descobrir se os nossos resíduos da regressão segue uma dstribuição normal. Faremos o teste Shapiro para avaliar a normalidade dos resíduos.

<center>$H_0\colon P_{\text{valor}} > 0.05$</center>
<center>$H_a\colon P_{\text{valor}} \leq 0.05$</center>

Caso P valor maior que 0.05, rejeita-se a hipótese alternativa e aceita a nula, onde a distribuição é normal. Caso P valor for menor ou igual a 0.05, rejeita-se a hipótese nula e e aceita a alternativa, onde a distribuição é diferente de uma normal.

```python
modelo1 = smf.ols('NOTA_REDACAO ~ COMP2 + COMP4 + COMP5', data = colegiox).fit()
residuos = modelo1.resid
```

```python
estatistica, p_valor = stats.shapiro(residuos)
p_valor
```

```python
stats.probplot(residuos, dist='norm',plot=plt);
```

Com o gráfico do QQ-Plot e o teste de Shapiro confirmando nossa análise, tendo o P-Valor maior que 0.05, confirmamos que os nossos resíduos possui distribuição normal, assegurando que o nosso modelo forneça bons resultados e seja preciso.

### 4. Teste da Homocedasticidade dos resíduos

O **teste da homocedasticidade** serve para **vermos** se a **variância de erros é constante entre os dados** (no caso seria os resíduos). Ela é desejável em modelos de regressões simples, pois permite fazer modelos mais confiáveis.


```python
plt.scatter(y=residuos, x=modelo1.predict(), color='red')
plt.hlines(y=0, xmin=0, xmax=1000, color='orange')
plt.xlabel('Resíduos')
plt.ylabel('Valores Preditos')
```

Mas, da mesma forma anterior do QQ-Plot, é bem dificil e subjetivo afirmar isso graficamente, vamos usar o teste de Breusch–Pagan, que nos confirma se nossos dados segue uma heterocedasticidade.
<center>$H_0\colon P_{\text{valor}} > 0.05$</center>
<center>$H_a\colon P_{\text{valor}} \leq 0.05$</center>

Sendo a hipótese nula a existência de homocedasticidade e a hipótese alternativa a não existência da homocedasticidade

```python
estatistica, p, f, fp = sms.het_breuschpagan(modelo1.resid, modelo1.model.exog)
print(f"O P valor é de {p}")
```

Com o P-Valor maior que 0.05, temos Homocedasticidade, ou seja, temos uma variação constante entre os dados.

### 5. Independência nos resíduos

Modelos que apresentam auto-correlação nos resíduos são claramente
identificados através da análise de resíduos, podendo ser identificado quando os resíduos não se comportam de forma aleatória, ou seja, seguem um padrão, e demonstram uma falha grave na especificação do modelo para o conjunto de dados.

Além da visualização gráfico, é possível validar se há auto-correlação nos resíduos, utilizando o teste estatístico chamado Durbin-Watson, que testa a hipótese nula de que os resíduos não são linearmente correlacionados automaticamente, o resultado final varia entre 0 e 4, e valores entre 1,5 <d <2,5 mostram que não há correlação.

```python
estatistica = statsmodels.stats.stattools.durbin_watson(modelo1.resid)
"Há independência nos resíduos" if ((estatistica > 1.5) and (estatistica < 2.5)) else "Não há independência nos resíduos"
```

## Implementação do modelo

Depois de ter feito todos os procedimentos para assegurar que o nosso modelo será bem aplicado e com bons resultados, iremos agora, efetivamente, realizar o modelo de regressão múltipla.

```python
print(modelo1.summary())
```

$\text{Equação:}$

<center>$\text{NOTA REDAÇÃO} = 49.6454 + \left[(\text{COMP1} \cdot 1.6047) + (\text{COMP4} \cdot 1.8418) + (\text{COMP5} \cdot 1.0578)\right]$</center>

$\text{Com R² ajustado de 0.968}$

```python
colegiox['Previsao'] = modelo1.fittedvalues
colegiox
```

### Métricas

#### Erro quadrático médio

$$\text{Erro Quadrático Médio} = \frac{1}{n} \cdot \sum\limits_{i=1}^{\mbox{n}}|( F_i \; - Y_i)^2|$$

$\text{Sendo:} \\ Y_i \text{ =  O valor real} \\ F_i \text{ = O valor previsto pelo modelo}$

```python
print(mean_squared_error(colegiox['NOTA_REDACAO'], colegiox['Previsao']))
```

### Erro médio absoluto

$$\text{Erro Médio Absoluto} = \frac{1}{n} \cdot \sum\limits_{i=1}^{\mbox{n}} |F_i \; - Y_i|$$

$\text{Sendo:} \\ Y_i \text{ =  O valor real} \\ F_i \text{ = O valor previsto pelo modelo}$

```python
print(mean_absolute_error(colegiox['NOTA_REDACAO'], colegiox['Previsao']))
```

### Raiz do erro quadrático médio
$$\text{Raiz do erro quadrático médio} = \sqrt{\frac{\sum\limits_{i=1}^{\mbox{n}}|( F_i \; - Y_i)^2|}{n}} $$

$\text{Sendo:} \\ Y_i \text{ =  O valor real} \\ F_i \text{ = O valor previsto pelo modelo}$

```python
print(np.sqrt(mean_squared_error(colegiox['NOTA_REDACAO'], colegiox['Previsao'])))
```
