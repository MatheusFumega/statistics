# Regressão Quantílica

A regressão quantílica nada mais é do que uma complementação da regressão linear. **Indicada**, principalmente, quando **existe** **outliers** e **heterocedasticidade**, isto é, quando os erros não são constantes ao longo de toda a amostra. Seu principal objetivo é estimar um determinado quantil da variável dependente. Exemplo: Equação obtida em um quantil de 25%

$$Y_4 = 10 \cdot X_4 + 0,5$$

Para um aumento de uma unidade em X, o quantil 25% (1 quartil) terá um aumento de 10 unidades.

Utiliza-se os quantis como referência, pois a média não descreve toda a distribuição, como no caso da regressão linear (justamente por ter tido falha nas premissas). 

A variável dependente precisa ser contínua.

## Validando o modelo Linear Simples

Precisamos ter em mente as premissas/suposições da regressão linear, caso contrário, o modelo falhará em fornecer bons resultados com os dados. Portanto, para um modelo bem-sucedido, é muito crucial validar essas suposições no conjunto de dados fornecido. Vamos tratar das seguintes premissas: 

**1.** O relacionamento entre a variável dependente e independente devem ser linear;
 
 
**2.** Não deve haver correlação entre as variáveis independentes;


**3.** Os resíduos devem ter uma distribuição normal;


**4.** Os resíduos devem ter variância constante (homocedasticidade), porém não deve ter uma heterocedasticidade para ser utilizada na quantílica.


**5.** Os resíduos não devem apresentar auto-correlação;


### 1. Linearidade entre variáveis independentes e dependentes

```python
sns.pairplot(dados);
```

```python
correlacoes = dados.corr(method='pearson')
plt.figure(figsize=(14, 11))
sns.heatmap(correlacoes, annot=True)
```

É possível perceber que há uma forte relação linear positiva entre a variável `dia` e as variáveis `num_refeicoes` e `gasto_acum_reais`.

Vamos utilizar a variável `dia` como **independente** e as demais como **dependente**.

### 2. Ausência de Multicolinearidade

Como temos somente uma variável independente, não é necessário avaliar a multicolinearidade entre variáveis independentes.

### 3. Normalidade dos resíduos

Para que a nossa regressão linear funcione de acordo, devemos descobrir se os nossos resíduos da regressão segue uma dstribuição normal. Faremos o teste Shapiro para avaliar a normalidade dos resíduos.

<center>$H_0\colon P_{\text{valor}} > 0.05$</center>
<center>$H_a\colon P_{\text{valor}} \leq 0.05$</center>

Caso P valor maior que 0.05, rejeita-se a hipótese alternativa e aceita a nula, onde a distribuição é normal. Caso P valor for menor ou igual a 0.05, rejeita-se a hipótese nula e e aceita a alternativa, onde a distribuição é diferente de uma normal.

```python
modelo1 = smf.ols('gasto_acum_reais ~ dia', data = dados).fit()
residuos = modelo1.resid
estatistica, p_valor = stats.shapiro(residuos)
print(f'Temos um P-Valor de: {p_valor}\nTemos a Estatística do teste de: {estatistica}')
```

```python
stats.probplot(residuos, dist='norm',plot=plt);
```

Temos um problema, com um P-Valor menor que 0.05 (4.715399882115889e-06) e um QQ-Plot não agradável, confirmamos uma distribuição não normal nos resíduos.

### 4. Teste da Homocedasticidade dos resíduos

O **teste da homocedasticidade** serve para **vermos** se a **variância de erros é constante entre os dados** (no caso seria os resíduos). Ela é desejável em modelos de regressões simples, pois permite fazer modelos mais confiáveis.


```python
plt.scatter(y=residuos, x=modelo1.predict(), color='red')
plt.hlines(y=0, xmin=0, xmax=2000, color='orange')
plt.xlabel('Resíduos')
plt.ylabel('Valores Preditos')
```

Mas, da mesma forma anterior do QQ-Plot, é bem dificil e subjetivo afirmar isso graficamente, vamos usar o teste de Breusch–Pagan, que nos confirma se nossos dados segue uma heterocedasticidade.
<center>$H_0\colon P_{\text{valor}} > 0.05$</center>
<center>$H_a\colon P_{\text{valor}} \leq 0.05$</center>

Sendo a hipótese nula a existência de homocedasticidade e a hipótese alternativa a não existência da homocedasticidade

```python
estatistica, p, f, fp = sms.het_breuschpagan(modelo1.resid, modelo1.model.exog)
print(f"O P valor é de {p}")
```

Com o P-Valor menor que 0.05, temos Heterocedasticidade, ou seja, rejeitamos a hipótese nula e confirmamos que há erros de resíduos não constantes ao longo de toda a amostra.

### 4.1 Teste de Outliers nos resíduos

```python
outliers = modelo1.outlier_test()
print(outliers.max())
print(outliers.min())
```

Usando uma análise entre 3 e -3, temos um outlier superior de 4.458

### 5. Independência nos resíduos

Modelos que apresentam auto-correlação nos resíduos são claramente
identificados através da análise de resíduos, podendo ser identificado quando os resíduos não se comportam de forma aleatória, ou seja, seguem um padrão, e demonstram uma falha grave na especificação do modelo para o conjunto de dados.

Além da visualização gráfico, é possível validar se há auto-correlação nos resíduos, utilizando o teste estatístico chamado Durbin-Watson, que testa a hipótese nula de que os resíduos não são linearmente correlacionados automaticamente, o resultado final varia entre 0 e 4, e valores entre 1,5 <d <2,5 mostram que não há correlação.

```python
estatistica = statsmodels.stats.stattools.durbin_watson(modelo1.resid)
"Há independência nos resíduos" if ((estatistica > 1.5) and (estatistica < 2.5)) else "Não há independência nos resíduos"
```

Temos graves problemas para utilizar esses dados em um modelo de Regressão Linear Simples:

- Ausência de normalidade
- Heterocedasticidade (Heterocedásticos)
- Outliers superiores
- Existência de auto-correlação nos resíduos

Portanto: **Modelo Reprovado** para Regressão Linear

## Modelo de Regressão Quantílica

Com o nosso modelo de regressão linear simples negado, temos outra alternativa de modelo: A **Regressão Quantílica**.

Para cada parâmetro na regressão quantílica, ou seja, para cada quantil que utilizamos, vamos percorrer a quantidade de dados em relação ao quantil.

```python
# Variável dependente gasto_acum_reais ; Variável independente dia
modelo1 = smf.quantreg('gasto_acum_reais ~ dia', dados)
resultado = modelo1.fit(q=0.25) # Utilizando essa faixa de valores
print(resultado.summary())
```

$\text{Equação para o quantil 0.25:}$

$$\text{Gasto Acumulado por dia} = -21,7143 + 30,9286 \cdot \text{Dia}$$

```python
# Variável dependente gasto_acum_reais ; Variável independente dia
modelo2 = smf.quantreg('gasto_acum_reais ~ dia', dados)
resultado2 = modelo2.fit(q=0.5) # Utilizando essa faixa de valores
print(resultado2.summary())
```

$\text{Equação para o quantil 0.5:}$

$$\text{Gasto Acumulado por dia} = -8,0313 + 30,7813 \cdot \text{Dia}$$

```python
# Variável dependente gasto_acum_reais ; Variável independente dia
modelo3 = smf.quantreg('gasto_acum_reais ~ dia', dados)
resultado3 = modelo3.fit(q=0.75) # Utilizando essa faixa de valores
print(resultado3.summary())
```

$\text{Equação para o quantil 0.75:}$

$$\text{Gasto Acumulado por dia} = 13,4074 + 30,4815 \cdot \text{Dia}$$

```python
# Variável dependente gasto_acum_reais ; Variável independente dia
modelo4 = smf.quantreg('gasto_acum_reais ~ dia', dados)
resultado4 = modelo4.fit(q=0.99) # Utilizando essa faixa de valores
print(resultado4.summary())
```

$\text{Equação para o quantil 0.99 (linear simples):}$

$$\text{Gasto Acumulado por dia} = -6,4127 + 32,4127 \cdot \text{Dia}$$

## Visualização gráfica dos modelos 

```python
plt_1 = plt.figure(figsize=(16, 12))
plt.scatter(x=dados['dia'], y=dados['gasto_acum_reais'], alpha=0.3)
X_plot = np.linspace(0, 64)
plt.xlabel("Dias")
plt.ylabel("Gasto Acumulado de reais")
plt.plot(X_plot, X_plot*resultado.params[1] + resultado.params[0], color='blue');
plt.plot(X_plot, X_plot*resultado2.params[1] + resultado2.params[0], color='red');
plt.plot(X_plot, X_plot*resultado3.params[1] + resultado3.params[0], color='green');
plt.plot(X_plot, X_plot*resultado4.params[1] + resultado4.params[0], color='orange');
plt.show();
```
