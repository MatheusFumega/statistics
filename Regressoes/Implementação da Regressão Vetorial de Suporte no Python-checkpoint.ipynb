# Regressão Vetorial de Suporte

- Aplicado a modelos lineares e não lineares.

- Para modelos não lineares, são utilizadas funções matemática (funções de kernel) não lineares (como polinômios).

- Ténica muito utilizada em algoritmo de Machine Learning (Máquina de Vetor de Suporte)

- O objetivo é encontrar um hiperplano que separa as classes.

$$f(x)=w \cdot x + b$$

Sendo:

$w = \text{Vetor peso}$

$x = \text{Vetor amostral}$

$b = \text{Constante}$

Graficamente:

![alt text for screen readers](vetor.png "Vetor")




(Classificar, separar de acordo com as classes com os dados que temos em mãos)

Os **Vetores de suporte** representam fronteiras, elementos que estão na fronteira de grupos que demonstram outra característica. A distância entre os dois vetores representa a **margem**. Dentro da **margem** é criado um **hiperplano**, que representa a separação entre os grupos com características específicas.

## Análise inicial dos dados

Vamos utilizar o banco de dados da **Donald Bren School of information and computer sciences**, classificaremos os tipos de plantas da família Iris, conforme suas dimensões de sépala e pétalas.

Descrição das variáveis:

- **`sepal-length`**: Comprimento da sépala
- **`sepal-width`**: Largura da sépala
- **`petal-length`**: Comprimento da pétala
- **`petal-width`**: Largura da pétala
- **`Class`**: O tipo dela

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.svm import SVC # Regressão Vetorial de Suporte para Classificação
from sklearn.svm import SVR # Regressão Vetorial de Suporte para Regressão
from sklearn.metrics import mean_absolute_error, mean_squared_error
import seaborn as sns
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

colnames = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'Class']

dados = pd.read_csv(url, names=colnames)
```

```python
dados
```

150 dados com 5 variáveis

## Separação dos atributos

```python
x = dados.drop('Class', axis=1) # Variáveis independentes
y = dados['Class'] # A variável dependente
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30) 
# Variáveis independentes, a variável dependente e o tamanho do teste (aleatório).
# Ele precisa saber se está certo ou não.
```

Desses 150 dados, iremos pegar uma **porcentagem** para usar como **treino** e uma **porcentagem** para usar como **teste** (30%).

## Modelo de Regressão Vetorial de Suporte: Linear

Existem quatro modelos vetorias de suporte: 

- Linear
- Sigmóide
- Polinomial
- Gaussiana

Será necessário testar cada uma para verificar qual modelo se encaixa melhor com os dados.

### Kernel = Linear

```python
svclassifier = SVC(kernel='linear')
svclassifier.fit(x_train, y_train)
```

```python
y_pred = svclassifier.predict(x_test)
```

```python
print(confusion_matrix(y_test, y_pred)) # Oque foi utilizado para test, qual foi a previsão obtida.
```

Uma comparação, para saber o quanto errou e o quanto acertou. A comparação é feia com o teste de dados que foi utilizado e o modelo de previsão.

A diagonal principal indica os acertos, os outros valores são os erros utilizando o Kernel Linear. As linhas e colunas representam, respectivamente, as classes que se encontram no banco de dados. Na última coluna (Virginica) e na segunda linha (Versicolor), é associado três erros a essas duas classes no modelo (O modelo pensou que fosse versicolor mas na realidade é virginica)

Linha (onde o modelo quis encaixar), Coluna (onde ele realmente pertence)

## Modelo de Regressão Vetorial de Suporte: Sigmóide


```python
svclassifier = SVC(kernel='sigmoid')
svclassifier.fit(x_train, y_train)
```

```python
y_pred = svclassifier.predict(x_test)
print(confusion_matrix(y_test, y_pred))
```

O modelo (com o Kernel Sigmóide) confundiu 15 Versicolors como Setosas, 16 Virgnicas como Setosas.

## Modelo de Regressão Vetorial de Suporte: Polinomial


```python
svclassifier = SVC(kernel='poly', degree=3)
svclassifier.fit(x_train, y_train)
```

```python
y_pred = svclassifier.predict(x_test)
print(confusion_matrix(y_test, y_pred))
```

## Modelo de Regressão Vetorial de Suporte: Gaussiana (rbf)


```python
svclassifier = SVC(kernel='rbf')
svclassifier.fit(x_train, y_train)
```

```python
y_pred = svclassifier.predict(x_test)
print(confusion_matrix(y_test, y_pred))
```

## Métricas do modelo

### Erro médio absoluto

$$\text{Erro Médio Absoluto} = \frac{1}{n} \cdot \sum\limits_{i=1}^{\mbox{n}} |F_i \; - Y_i|$$

$\text{Sendo:} \\ Y_i \text{ =  O valor real} \\ F_i \text{ = O valor previsto pelo modelo}$

```python
print(mean_absolute_error(y, previsoes)) # Y_true = Variável dependente, Y_pred = Previsão da variável dependente

# ou 

colegiox['erro_abs_medio'] = abs(previsoes - y['NOTA_REDACAO'])

print(colegiox['erro_abs_medio'].mean())
```

### Erro quadrático médio

$$\text{Erro Quadrático Médio} = \frac{1}{n} \cdot \sum\limits_{i=1}^{\mbox{n}}|( F_i \; - Y_i)^2|$$

$\text{Sendo:} \\ Y_i \text{ =  O valor real} \\ F_i \text{ = O valor previsto pelo modelo}$

```python
print(mean_squared_error(y, previsoes))

# ou

colegiox['erro_quadratc_medio'] = abs(previsoes - y['NOTA_REDACAO'])**2

colegiox['erro_quadratc_medio'].mean()
```

### Raiz do erro quadrático médio
$$\text{Raiz do erro quadrático médio} = \sqrt{\frac{\sum\limits_{i=1}^{\mbox{n}}|( F_i \; - Y_i)^2|}{n}} $$

$\text{Sendo:} \\ Y_i \text{ =  O valor real} \\ F_i \text{ = O valor previsto pelo modelo}$

```python
print(np.sqrt(mean_squared_error(y, previsoes)))
```

## Regressão Vetorial de Suporte com Escalonamento

E se as nossas variáveis independentes forem longe da escala? Por exemplo, o nosso componente `COMP1` for 5, o `COMP2` for 1000 e o `COMP4` for 32000, são numeros com escalas diferentes, uma com dezena a outra com milhar, isso dará um grande problema para a Máquina Vetorial de Suporte. Visto isso, podemos usar o **escalonamento**, colocar todos os valores na mesma escala.

### SVR com escalonamento
Vamos utilizar o método da Padronização de escala.

```python
from sklearn.preprocessing import StandardScaler
escalax = StandardScaler() # Nossa padronização

independente_escala = escalax.fit_transform(x) # Padronizar os valores da variável idependente (COMP1, COMP4, COMP5)
```

Será transformado de tal maneira que, a média será próximo a zero e o desvio padrão próximo de 1 (Padronização).

```python
independente_escala
```

```python
escalay = StandardScaler() # Nossa padronização

dependente_escala = escalay.fit_transform(y) # Padronizar os valores da variável dependente (NOTA_REDACAO)
```

```python
dependente_escala
```

Os valores ficaram na mesma escala, com média próxima de 0 e desvio padrão igual a 1. Para esse modelo de regressão, quando existe uma diferença de valores muito grande, é extremamente importante que exista essa padronização

```python
SVR2 = SVR(kernel='linear')
SVR2.fit(independente_escala, dependente_escala.ravel())
```

```python
SVR2.score(independente_escala, dependente_escala)
```

Ficou um pouco pior comparado antes da padronização que utilizamos (0.9230695631462075)

```python
SVR2.predict(independente_escala)
```
