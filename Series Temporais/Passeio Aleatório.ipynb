```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6
from random import sample, random
```

```python
dados = sample(range(100), k=41) # 41 Valores, entre 0 e 100
dados
```

```python
anos = list(range(1980,2021))
anos
```

```python
serie = pd.Series(dados, index = anos)
```

```python
serie.plot()
plt.show() # Sem coer√™ncia de repeti√ß√£o, um passeio aleat√≥rio.
```

```python
serie2 = list()
serie2.append(-1 if random() < 0.5 else 1)
print(serie2)
for i in range(1 , 1000):
    movimento = -1 if random() < 0.5 else 1
    valor = serie2[i-1] + movimento #
    serie2.append(valor)
    print(f'O movimento ser√° {movimento} \n O valor ser√° {serie2[i-1]} + {movimento}')
    print(serie2)
```

```python
serie2 = pd.Series(serie2)
```

```python
serie2.plot()
plt.show()
```
